```python
%matplotlib inline
import scipy.stats as stats
from matplotlib import pyplot as plt
```

```python
npdf = stats.norm(0,10)
plt.plot(range(-100,100), npdf.pdf(range(-100,100)), 'k-', lw=2)
sum(npdf.pdf(range(-100, 100)))
npdf.cdf(60)

hist=plt.hist(npdf.rvs(1000), bins=200, normed=True)
```

```python
poi = stats.poisson
random_vars = poi.rvs(15, size=1000)
hist=plt.hist(random_vars, bins=20, normed=True)
x = range(0,100)
plt.plot(x, poi.pmf(x, mu=30), 'k-', lw=2)
```
